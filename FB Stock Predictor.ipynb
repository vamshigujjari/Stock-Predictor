```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
training_set = pd.read_csv('FB_train.csv')
```

```python
training_set.drop('Adj Close', axis = 1)
```

```python
training_set = training_set.iloc[:,1:2].values
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
sc = MinMaxScaler()
```

```python
training_set = sc.fit_transform(training_set)
```

```python
training_set
```

```python
X_train = training_set[0:1240]
```

```python
X_train
```

```python
Y_train = training_set[1:1241]
```

```python
X_train = np.reshape(X_train, (1240, 1, 1))
```

```python
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
regressor = Sequential()
```

```python
regressor.add(LSTM(units = 4, activation = 'sigmoid', input_shape = (None, 1)))
```

```python
regressor.add(Dense(units = 1))
```

```python
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
regressor.fit(X_train, Y_train, batch_size = 32, epochs = 150)
```

```python
test_set = pd.read_csv('FB_test.csv')
real_stock = test_set.iloc[:,1:2].values
```

```python
real_stock
```

```python
inputs = real_stock
inputs.shape
```

```python
inputs = sc.transform(inputs)
```

```python
inputs.shape
```

```python
inputs = np.reshape(inputs,(20,1,1))
```

```python
predicted_price = regressor.predict(inputs)
```

```python
predicted_price = sc.inverse_transform(predicted_price)
```

```python
predicted_price
```

```python
plt.plot(real_stock, color = 'red', label = 'Real FB Stock')
plt.plot(predicted_price, color = 'blue', label = 'Predicted FB Stock')
plt.title('FB Stock Price Prediction')
plt.xlabel('Time')
plt.ylabel('FB Stock Price')
plt.legend()
plt.show()
```

```python
import math
from sklearn.metrics import mean_squared_error
```

```python
rmse = math.sqrt(mean_squared_error(real_stock, predicted_price))
rmse
```
